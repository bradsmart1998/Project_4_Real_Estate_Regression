## Nearest School

```python
#Count number of schools
near_school = housing_df["NEAREST_SCH"].value_counts()
near_school
```

```python
# How many unique NEAREST_SCH are there in this dataset
housing_df["NEAREST_SCH"].nunique()
```

##### Swan View Senior high school is the nearest school to most of the houses in several different suburbs. 


```python
top_20_sch = (housing_df['NEAREST_SCH'].value_counts()).iloc[:20]
top_20_sch
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
top_20_sch.plot(kind="bar", figsize=(20,3.5))

# Set a title for the chart
plt.title("Number of addresses near each school")

plt.show()
plt.tight_layout()
```

```python
# # Create a bar chart to show number of addresses each school is nearest too -for all
near_sch_chart = near_school.plot(kind='bar', figsize=(30,8))

# # Set the xlabel and ylabel using class methods
near_sch_chart.set_xlabel("Names of Schools")
near_sch_chart.set_ylabel("Number of nearest addresses")

plt.show()
plt.tight_layout()
```

```python
# export to data to csv
os.makedirs('folder/subfolder', exist_ok=True)  
near_school.to_csv('folder/subfolder/Nearest_schools.csv')  
```

### NEAREST SCH RANK

```python
housing_df["NEAREST_SCH_RANK"].nunique()
```

```python
housing_df["NEAREST_SCH_RANK"].value_counts()
```

```python
#Create boxplot for NEAREST_SCH_RANK, where NaN values are present
plt.figure(figsize=(10,8))
sns.boxenplot(x=housing_df["NEAREST_SCH_RANK"])
```

```python
#calculate the mean for nearest sch rank
mean_rank = housing_df["NEAREST_SCH_RANK"].mean()
```

```python
mean_rank
```

```python
#replace NaN value with mean value
replace_nan_rank = housing_df['NEAREST_SCH_RANK'] = housing_df['NEAREST_SCH_RANK'].fillna(mean_rank)
```

```python
replace_nan_rank
```

```python
#Create boxplot for NEAREST_SCH_RANK, where NaN values are replaced by mean value
plt.figure(figsize=(10,8))
sns.boxenplot(x=replace_nan_rank)
```

```python
#nunique after the Nan replacement
housing_df["NEAREST_SCH_RANK"].nunique()
```

```python
#value count of the nearest sch rank after the NaN reaplcement
housing_df["NEAREST_SCH_RANK"].value_counts()
```

```python
#groupby the suburb and nearest school rank
# First grouping based on "SUBURB"
# Within each SUBURB we are grouping based on "NEAREST SCH RANK"
school_rank = housing_df.groupby(['SUBURB', 'NEAREST_SCH_RANK'])
```
